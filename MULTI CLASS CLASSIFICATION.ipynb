An imputing method should be chosen to fill missing values. Options are:


1. Simple Imputation (Mean or Median)
2. MICE Imputation
3. kNN Imputation
4. Simply dropping NaN values

Becayse there is too many null values in the dataset, it would be inefficient to simple impute or drop the values. MICE imputation is better choice than kNN imputation because data is not distributed ordinally.

Thus, MICE imputation will be used.



```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
mice_imputer = IterativeImputer(random_state=100, max_iter=30)
mice_imputer.fit(data) #takes a while
```

```python
data_mice = mice_imputer.transform(data)
data_mice = pd.DataFrame(data_mice, columns = data.columns)
```

```python
data_mice.isnull().sum()
```

Compute Euclidian values for x, y, and z to extract features and reduce dimensions

```python
def add_euclidian(X,Y,Z,E,df):
  df[E] = np.sqrt(df[X]**2 + df[Y]**2 + df[Z]**2)
  df = df.drop([X, Y, Z], axis = 1)
  return df
```

```python
data_mice = add_euclidian('X0', 'Y0', 'Z0', 'E0', data_mice)
data_mice = add_euclidian('X1', 'Y1', 'Z1', 'E1', data_mice)
data_mice = add_euclidian('X2', 'Y2', 'Z2', 'E2', data_mice)
data_mice = add_euclidian('X3', 'Y3', 'Z3', 'E3', data_mice)
data_mice = add_euclidian('X4', 'Y4', 'Z4', 'E4', data_mice)
data_mice = add_euclidian('X5', 'Y5', 'Z5', 'E5', data_mice)
data_mice = add_euclidian('X6', 'Y6', 'Z6', 'E6', data_mice)
```

'User' column is dropped since model performs better without it.

```python
data_mice = data_mice.drop(['User'], axis = 1)
```

# Model Selection and Evaluation

1 Linear and 2 non - linear models are chosen for this data.

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import confusion_matrix
```

```python
def fit_evaluate(X, y, model_call):

  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=123)

  model = model_call.fit(X_train, y_train)
  y_pred = model.predict(X_test)

  train_acc = model.score(X_train, y_train)
  test_acc = model.score(X_test, y_test)

  precision, recall, f1_score, support = precision_recall_fscore_support(y_test, y_pred)

  scores = [train_acc, test_acc, np.mean(precision), np.mean(recall), np.mean(f1_score), np.mean(support)]

  print(f"Mean train accuracy for {model_call} is : {train_acc}")
  print(f"Mean test accuracy for {model_call} is : {test_acc}")
  print(f"Mean precision for {model_call} is : {np.mean(precision)}")
  print(f"Mean recall for {model_call} is : {np.mean(recall)}")
  print(f"Mean F1 score for {model_call} is : {np.mean(f1_score)}")
  print(f"Mean support value for {model_call} is : {np.mean(support)}")

  cm = confusion_matrix(y_test, y_pred)
  plt.figure(figsize=(12, 8))
  sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=model.classes_, yticklabels=model.classes_)
  plt.title(f'Confusion Matrix - {model_call}')
  plt.xlabel('Predicted Labels')
  plt.ylabel('True Labels')
  plt.show()

  return scores
```

SVM

```python
scores_svm = fit_evaluate(X, y, make_pipeline(StandardScaler(), SVC(gamma='auto'))) #takes a good while
```

Naive Bayes Classifier

```python
scores_naiv = fit_evaluate(X, y, GaussianNB())
```

Random Forest Classifier

```python
scores_rf = fit_evaluate(X, y, RandomForestClassifier(max_depth=None, random_state=123))
```

Compare the performances

```python
def generate_metrics(scores):
  metrics = []
  for i in range(6):
    arr = [model_scores[i] for model_scores in scores]
    metrics.append(arr)
  return metrics
```

```python
scores = [scores_svm, scores_naiv, scores_rf]
metrics = generate_metrics(scores)
models = ['SVM', 'Naive Bayes', 'Random Forest']

fig, axes = plt.subplots(2, 3, figsize=(12, 6))


sns.barplot(x=models, y=metrics[0], ax=axes[0, 0])
axes[0, 0].set_title(f'Mean Train Accuracies')
axes[0, 0].set_xlabel('Models')
axes[0, 0].set_ylabel('Accuracy')

sns.barplot(x=models, y=metrics[1], ax=axes[0, 1])
axes[0, 1].set_title(f'Mean Test Accuracies')
axes[0, 1].set_xlabel('Models')
axes[0, 1].set_ylabel('Accuracy')

sns.barplot(x=models, y=metrics[2], ax=axes[0, 2])
axes[0, 2].set_title(f'Mean Precisions')
axes[0, 2].set_xlabel('Models')
axes[0, 2].set_ylabel('Precision')

sns.barplot(x=models, y=metrics[3], ax=axes[1, 0])
axes[1, 0].set_title(f'Mean Recalls')
axes[1, 0].set_xlabel('Models')
axes[1, 0].set_ylabel('Recall')

sns.barplot(x=models, y=metrics[4], ax=axes[1, 1])
axes[1, 1].set_title(f'mean F1 Scores')
axes[1, 1].set_xlabel('Models')
axes[1, 1].set_ylabel('Scores')

sns.barplot(x=models, y=metrics[5], ax=axes[1, 2])
axes[1, 2].set_title(f'mean Support Values')
axes[1, 2].set_xlabel('Models')
axes[1, 2].set_ylabel('Value')
+


plt.tight_layout()
plt.show()
```
